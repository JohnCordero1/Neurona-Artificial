# Algunas funciones de activación

## Función Escalón (Heaviside)




Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = \begin{cases}
        0, & \text{si } z < 0 \\
1, & \text{si } z \geq 0
\end{cases}$$

Donde el rango de esta función esta acotada, $φ(z) \in \left\{0,1 \right\}$ y, la derivada no existe:

$$\nexists \frac{d }{d x} \hat{y}$$

```python
# Definir la función escalón
def escalon(x):
    return np.heaviside(x, 1)

# Generar datos para la función escalón
x1 = np.linspace(-5, 5, 100)
y1 = escalon(x1)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, label='Función Escalón', color ="red", linewidth=2.5)
plt.title('Función Escalón (Heaviside)', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

## Función Lineal (identidad)



Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = z$$

Donde el rango de esta función no esta acotada, $φ(z) \in (- \infty, \infty )$ y, la derivada es:

$$\frac{d}{d x}\hat{y} = \frac{d \varphi(z)}{dx} = 1$$

```python
# Definir la función lineal
def funcion_lineal(x, m, b):
    return m * x + b

# Generar datos para x (por ejemplo, de -10 a 10)
x2 = np.linspace(-5, 5, 100)

# Definir los parámetros de la función lineal
pendiente = 1
ordenada_al_origen = 0

# Calcular los valores de y utilizando la función lineal
y2 = funcion_lineal(x2, pendiente, ordenada_al_origen)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x2, y2, label='Función Lineal', color='blue', linewidth= 2.5)
plt.title('Función Lineal (Identidad)')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()
```

## Función Logística

Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = \frac{1}{1- e^{-z}}$$

Donde el rango de esta función esta, $φ(z) \in (0, 1)$ y, la derivada es:

$$\frac{d}{d x}\hat{y} = \frac{d \varphi(z)}{dx} = \hat{y}(1-\hat{y})$$

```python
# Definir la función logística
def logistic_function(x):
    return 1 / (1 + np.exp(-x))

# Crear datos para el eje x
x3 = np.linspace(-5, 5, 100)

# Calcular los valores de la función logística
y3 = logistic_function(x3)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x3, y3, label='Función Logística', color='purple', linewidth=2.5)
plt.title('Función Logística')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend()
plt.show()
```

## Función Tangente Hiperbólica

Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = \text{tanh}(z)$$

Donde el rango de esta función esta, $φ(z) \in (-1, 1)$ y, la derivada es:

$$\frac{d}{d x}\hat{y} = \frac{d \varphi(z)}{dx} = (1+\hat{y})(1-\hat{y})$$

```python
# Definir la función tangente hiperbólica
def tanh_function(x):
    return np.tanh(x)

# Crear datos para el eje x
x4 = np.linspace(-5, 5, 100)

# Calcular los valores de la función tangente hiperbólica
y4 = tanh_function(x4)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x4, y4, label='Función Tangente Hiperbólica', color='green', linewidth=2.5)
plt.title('Función Tangente Hiperbólica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Función Softplus

Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = \ln (1+e^z)$$

Donde el rango de esta función esta, $φ(z) \in (0, \infty)$ y, la derivada es:

$$\frac{d}{d x}\hat{y} = \frac{d \varphi(z)}{dx} = \frac{1}{1+e^{-z}}$$

```python
# Definir la función Softplus
def softplus_function(x):
    return np.log(1 + np.exp(x))

# Crear datos para el eje x
x5 = np.linspace(-5, 5, 100)

# Calcular los valores de la función Softplus
y5 = softplus_function(x5)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x5, y5, label='Función Softplus', color='orange', linewidth=2.5)
plt.title('Función Softplus')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Función ReLU (Rectified Linear Unit)

Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = \max (0,z)$$

Donde el rango de esta función esta, $φ(z) \in [0, \infty)$ y, la derivada es:

$$ \frac{d}{d x}\hat{y} ≜ \begin{cases}
        0, & \text{si } z < 0 \\
1, & \text{si } z \geq 0
\end{cases}$$

```python
# Definir la función ReLU
def relu_function(x):
    return np.maximum(0, x)

# Crear datos para el eje x
x6 = np.linspace(-5, 5, 100)

# Calcular los valores de la función ReLU
y6 = relu_function(x6)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x6, y6, label='Función ReLU', color='brown', linewidth=2.5)
plt.title('Función ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Función Leaky ReLU (Rectified Linear Unit)

Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = \begin{cases}
        0.01z, & \text{si } z < 0 \\
z, & \text{si } z \geq 0
\end{cases}$$

Donde el rango de esta función esta, $φ(z) \in (- \infty, \infty)$ y, la derivada es:

$$ \frac{d}{d x}\hat{y} ≜ \begin{cases}
        0.1, & \text{si } z < 0 \\
1, & \text{si } z \geq 0
\end{cases}$$

```python
# Definir la función Leaky ReLU
def leaky_relu_function(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Crear datos para el eje x
x7 = np.linspace(-5, 5, 100)

# Calcular los valores de la función Leaky ReLU
y7 = leaky_relu_function(x7)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x7, y7, label='Función Leaky ReLU', color='blue', linewidth=2.5)
plt.title('Función Leaky ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Función Gaussiana

Si tenemos la siguiente combinación lineal $z = w^T x + b $, con la siguiente función de activación:

$$ \hat{y} = φ(z) = e ^{-x^{2}}$$

Donde el rango de esta función esta, $φ(z) \in (0, 1]$ y, la derivada es:

$$\frac{d}{d x}\hat{y} = \frac{d \varphi(z)}{dx} = -2z\hat{y}$$

```python
# Definir la función gaussiana
def gaussian_function(x, mu, sigma):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

# Crear datos para el eje x
x8 = np.linspace(-5, 5, 100)

# Definir la media y la desviación estándar
mu = 0
sigma = 1

# Calcular los valores de la función gaussiana
y8 = gaussian_function(x8, mu, sigma)

# Crear el gráfico
plt.figure(figsize=(10, 5))
plt.plot(x8, y8, label='Función Gaussiana', color='purple', linewidth=2.5)
plt.title('Función Gaussiana')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Regla Delta

Podemos generalizar todas las funciones de activación a una sola regla, los conceptos sobre que es una función de activación no tiene una definición cerrada sobre lo que es y lo que no es; sin embargo hay algunas características que podemos observar de estas funciones de activación:

- Conocimiento sobre si la función de activación es lineal o no lineal.
- Rango de la función.
- Si la función es diferenciable.
- Monótonamente creciente o decreciente (importante para que la función de costo sea convexa).

Entonces, si tenemos la siguiente función de costo para el entrenamiento en linea:

$$L(y,\hat{y}) = \frac{1}{2}(y-\hat{y})^2$$

Y queeremos aplicar el algoritmo de optimización del gradiente descendente:

\begin{align}
w - η ∇_w L(y,\hat{y}) \rightarrow w \\
b - η \frac{∂}{∂ b}  L(y,\hat{y}) \rightarrow b \\
\end{align}

Las ecuaciones de actualización serán:

\begin{align}
w + η (y-\hat{y}) \left(\frac{d \hat{y}}{d z}\right) x^T \rightarrow w \\
b + η (y-\hat{y}) \left(\frac{d \hat{y}}{d z}\right) \rightarrow b \\
\end{align}

Por lo tanto, solo debemos adaptar las ecuaciones de actualización según las derivadas que de las funciones de activación.

Para el caso de que tengamos el error cuadratico medio o entrenamiento en lote:

$$L(y, \hat{y}) = \frac{1}{2p} \sum_{i=1}^p(y^{(i)} - \hat{y}^{(i)})^2$$

El algoritmo de optimización es:

\begin{align}
w - η ∇_w L(y,\hat{y}) \rightarrow w \\
b - η \frac{∂}{∂ b}  L(y,\hat{y}) \rightarrow b \\
\end{align}

Y las ecuaciones de actualización serán:

\begin{align}
w + \frac{η}{p} \left[\left(Y-\hat{Y}\right) \odot \left(\frac{d\hat{Y}}{dZ}\right)\right] X^T \rightarrow w \\
b + \frac{η}{p} \sum \left[\left(Y-\hat{Y}\right) \odot \left(\frac{d\hat{Y}}{dZ}\right)\right] \rightarrow b \\
\end{align}

# Programando la Regla Delta

```python
#Creamos las funciones de activacion con su derivada
def linear(z, derivative=False):
  a = z
  #Calculamos la derivada de la funcion lineal
  if derivative:
    da = np.ones(z.shape) #vector de unos de las misma forma que z
    return a, da
  return a


def sigmoid(z, derivative=False):
  a = 1 / (1 + np.exp(-z))
  if derivative:
    da = a * (1 - a) #Sera producto elemento a alemento de un dos vectores
    return a, da
  return a


def tanh(z, derivative=False):
  a = np.tanh(z)
  if derivative:
    da = (1+a) * (1-a)
    return a, da
  return a


def relu(z, derivative=False):
  a = z * (z >= 0) #Devolvera vector de True o False, multiplicamos por z para obtener los True
  if derivative:
    da = np.array(z >= 0, dtype=np.float32) #vector de 1
    return a, da
  return a


#Creamos la clase con el numero de entradas, la funcion de activacion (por defecto la lineal) y el factor de aprendizaje
class neuron:
  def __init__(self, n_inputs, act_fun=linear, learning_rate=0.1):
    self.w = -1+2*np.random.rand(n_inputs)
    self.b = -1+2*np.random.rand()
    self.eta = learning_rate
    self.f = act_fun

  #Creamos el metodo de prediccion
  def predict(self, X):
    Z = np.dot(self.w, X) + self.b
    return self.f(Z) #Retornamos la funcion de activación

  #Creamos el metodo para el entrenamiento
  def fit(self, X, Y, epochs=100):
    #Numero de patrones necesarios, de acuerdo a la segunda dimension de X
    p = X.shape[1]
    #Implementamos el error cuadratico medio
    for _ in range(epochs):
      #Propagacion
      Z = np.dot(self.w, X) + self.b
      Yest, dy = self.f(Z, derivative=True)
      #Entrenamiento
      lg = (Y - Yest) * (dy)
      self.w += (self.eta/p) * np.dot(lg, X.T).ravel() #ravel para que quede en una dimension
      self.b += (self.eta/p) * np.sum(lg)
```

```python
#Ejemplos
example = 1

if example == 1:
  X = np.array([[0,0,1,1],
                [0,1,0,1]])
  Y = np.array([[0,0,0,1]])

  #instanciamos con el nombre de neurona
  neurona = neuron(2, sigmoid, 0.9)
  neurona.fit(X,Y, epochs=500)

  p = X.shape[1]
  for i in range(p):
    if Y[0,i] == 0:
      plt.plot(X[0,i], X[1,i], "ob")
    else:
      plt.plot(X[0,i], X[1,i], "or")

  w1, w2, b = neurona.w[0], neurona.w[1], neurona.b
  plt.plot([-2,2], [(1/w2)*(-w1*(-2)-b), (1/w2)*(-w1*2-b)], "--k")

  plt.xlim([-1,2])
  plt.ylim([-1,2])
  plt.xlabel("X_1")
  plt.ylabel("X_2")

else:
  p = 100
  x = -1+2*np.random.rand(p).reshape(1,-1)
  y = -18*x+6+2.5*np.random.randn(p)
  plt.plot(x,y,".b")

  #instanciamos con el nombre de neurona
  neurona = neuron(1, linear, 0.1)
  neurona.fit(x,y, epochs=100)

  xn = np.array([[-1,1]])
  plt.plot(xn.ravel(), neurona.predict(xn), "--r")
```
